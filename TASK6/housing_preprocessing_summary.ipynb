```python
import pandas as pd
```

```python
# Read housing.csv file here
try:
  housing_df = pd.read_csv('california_housing_data.csv')
except:
  print("ERROR: could not read in california_housing_data.csv.")

print(housing_df)
housing_df.size
```

```python
# Read california_housing_data.csv file here
try:
  ca_cities_df = pd.DataFrame(pd.read_csv("cal_cities_lat_long.csv"))
except:
  print("ERROR: could not read in cal_cities_lat_long.csv.")

ca_cities_df.head(10)
```

```python
#convert the latitude and longitude on housing_df dataframe to 2 decimal precision
housing_df['latitude'] = round(housing_df['latitude'],2)
housing_df['longitude'] = round(housing_df['longitude'],2)
print(housing_df.loc[:,['latitude', 'longitude']])
housing_df.size
```

```python
#convert the latitude and longitude on ca_cities_df dataframe to 2 decimal precision
ca_cities_df['latitude'] = round(ca_cities_df['Latitude'],2)
ca_cities_df['longitude'] = round(ca_cities_df['Longitude'],2)
ca_cities_df = ca_cities_df.drop(['Latitude', 'Longitude'],axis=1)
print(housing_df.loc[:,['latitude', 'longitude']])
housing_df.size
```

```python
# Join 2 data frames using left outer join.
joined_df = pd.merge(ca_cities_df,housing_df, on=['latitude', 'longitude'], how='left', indicator=True)
print(joined_df.head(10))
joined_df.size
```

```python
# Find number of rows missing the city information
# That is find out the Name column with "NaN" using isna() function and then do
# sum() function on top of that dataframe
print("The total number of rows missing the city information: ", joined_df['Name'].isna().sum())
```

```python
# Now select only the rows which has city populated
# That means drop any rows which has NaN value in the cell
filtered_joined_df = joined_df.dropna(subset=['Name'], axis=0)
filtered_joined_df
```

```python
#Find the minimum and maximum housing_median_age
print("Minimum Housing Median Age is: ", filtered_joined_df['housing_median_age'].min())
print("Maximum Housing Median Age is: ", filtered_joined_df['housing_median_age'].max(), "\n")

#Find the minimum and maximum total_bedrooms
print("Minimum Total Bedrooms is: ", filtered_joined_df['total_bedrooms'].min())
print("Maximum Total Bedrooms is: ", filtered_joined_df['total_bedrooms'].max(), "\n")

#Find the minimum and maximum median_income
print("Minimum Median Income is: ", filtered_joined_df['median_income'].min())
print("Maximum Median Income is: ", filtered_joined_df['median_income'].max(), "\n")

#Find the minimum and maximum median_house_value
print("Minimum Median House Value is: ", filtered_joined_df['median_house_value'].min())
print("Maximum Median House Value is: ", filtered_joined_df['median_house_value'].max())
```

```python
#Now select the rows with housing_median_age less than 25 year and its value less than 100000

final_df = filtered_joined_df.query('housing_median_age < 25 and median_house_value < 100000')
print(final_df)
final_df.size
```
